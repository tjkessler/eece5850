```python
##############################
### Exact prime generation ###
##############################

from rsa.keygen import generate_key
from time import time
from matplotlib import pyplot as plt

times = []
bits = []
for i in range(4, 48):
    _t_start = time()
    _, _ = generate_key(i)
    times.append(time() - _t_start)
    bits.append(i)

plt.clf()
plt.rcParams['font.family'] = 'Times New Roman'
plt.xlabel('Number of bits in $p$ and $q$')
plt.ylabel('Time to generate public and private keys (s)')
plt.plot(bits, times, color=(0, 114/235, 178/235))
plt.show()
```

```python
################################################
### Miller-Rabin method for prime generation ###
################################################

from rsa.miller_rabin import generate_key as generate_mr

times_mr = []
bits = []
for i in range(4, 256):
    _t_start = time()
    _, _ = generate_mr(i)
    times_mr.append(time() - _t_start)
    bits.append(i)

plt.clf()
plt.rcParams['font.family'] = 'Times New Roman'
plt.xlabel('Number of bits in $p$ and $q$')
plt.ylabel('Time to generate public and private keys (s)')
plt.plot(bits, times_mr, color=(213/235, 94/235, 0))
plt.show()
```

```python
# Direct comparison

plt.clf()
plt.rcParams['font.family'] = 'Times New Roman'
plt.xlabel('Number of bits in $p$ and $q$')
plt.ylabel('Time to generate public and private keys (s)')
plt.plot(bits[:25], times[:25], color=(0, 114/235, 178/235), label='Exact prime generation')
plt.plot(bits[:25], times_mr[:25], color=(213/235, 94/235, 0), label='Miller-Rabin generation')
plt.legend(loc='upper left')
plt.show()
```

```python
t_start = time()
_, _ = generate_mr(512, k=40)
print('Time to generate 512-bit public and private keys: {:.5f} seconds'.format(time() - t_start))
```

```python
times_k = []
ks = []
for k in range(256):
    _t_start = time()
    _, _ = generate_mr(64, k=k)
    times_k.append(time() - t_start)
    ks.append(k)

plt.clf()
plt.rcParams['font.family'] = 'Times New Roman'
plt.xlabel('Number of Miller-Rabin tests $k$')
plt.ylabel('Time to generate 64-bit key (s)')
plt.plot(ks, times_k, color=(0, 158/235, 115/235))
plt.show()
```
